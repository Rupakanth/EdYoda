### Q1. Add average marks section ?

```python
def f(r):
    return (r.G1 + r.G2)/2

def add_avg_marks(data):
    data['average_marks'] = data.apply(f,axis = 1)
    return data
```

### Q13. Find score, test for few more models (LogisticRegression, DecisionTree, RandomForest )

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
def testing_alogorithms(trainX,trainY,testX,testY):

    #test for logistic regression algorithm
    log_reg = LogisticRegression()
    log_reg.fit(trainX,trainY)
    print('Logistoc Regression Score:',log_reg.score(testX,testY))
    
    
    #test for DecisionTreeClassifier algorithm
    dt = DecisionTreeClassifier()
    dt.fit(trainX,trainY)
    print("Decision Tree score:",dt.score(testX,testY))
    
    #test for RandomForestClassifier algorithm
    rf = RandomForestClassifier()
    rf.fit(trainX,trainY)
    print('Random Forest Score:',rf.score(testX,testY))
```

```python
testing_alogorithms(trainX,trainY,testX,testY)
```

### Q14. Cluster the students into 3 groups using KMeans

```python
from sklearn.cluster import KMeans
```

```python
def seperate_in_clusters(df):
    kmeans = KMeans(n_clusters=3)
    kmeans.fit(df)
    
    print('Cluster centers :', kmeans.cluster_centers_,'\n', end='\n')
    print('Cluster labels :', kmeans.labels_)
    
    return kmeans
```

```python
seperate_in_clusters(df)
```
