```python
missing_values = ['na','HURLEY','--','12','n/a','1.5']
property_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/Data-Scientist-program/master/Practice%20Problems/data/property%20data.csv', na_values = missing_values)
property_data
```

Filling the missing value in PID with 100005000.0 since data is given in sequence and 100005000.0 is missing

```python
property_data['PID'].fillna(100005000.0,inplace = True)
```

```python
property_data
```

filling missing value if ST_NUM of LEXINGTON as 199.0,since street numbers of same ST_NAME vary by 2
and WASHINGTON by 210.0

```python
property_data[2:3]['ST_NUM'] = 199.0
property_data[6:7]['ST_NUM'] = 210.0
```

```python
property_data
```

since in BERKELEY most of the houses are own occupied,filling the NaN cell as Y 
and for washington filling it with the mode of the column which is Y

```python
property_data[3:4]['OWN_OCCUPIED'] = 'Y'
```

```python
property_data['OWN_OCCUPIED'].fillna(property_data['OWN_OCCUPIED'].mode()[0],inplace = True)
property_data
```

```python
# in data NUM_BATH for PID = 100002000.0 is 1.5 which might be a data entry error,rpalcong it with 1.0 and the SQ_FT as 1000
property_data[1:2]['NUM_BATH'] = 1.0
property_data[1:2]['SQ_FT'] = 1000.0


# in data for PID = 100003000.0 is NUM_BATH = 1 and SQ_FT = 850, from observation guessing NUM_BEDROOMS = 2
property_data[2:3]['NUM_BEDROOMS'] = 2.0

# in data for PID = 100004000.0 is NUM_BEDROOMS = 1 and SQ_FT = 700, from observation guessing NUM_BATH = 1.for 700SQ_FT
property_data[3:4]['NUM_BATH'] = 1.0

# in data for PID = 100006000.0 is  NUM_BATH = 1 and SQ_FT = 800, from observation guessing NUM_BEDROOMS = 1.for 800SQ_FT
property_data[5:6]['NUM_BEDROOMS'] = 1.0

# in data for PID = 100007000.0 is  NUM_BEDROOMS = 2 and SQ_FT = 950, from observation guessing NUM_BATH = 2.for 950SQ_FT
property_data[6:7]['NUM_BATH'] = 2.0

# in data for PID = 100008000.0 is  NUM_BEDROOMS = 1 and NUM_BATH = 2, from observation guessing SQ_FT = 900
property_data[7:8]['SQ_FT'] = 900.0

# in data for PID = 100009000.0 is  NUM_BATH = 2 and SQ_FT = 1800, from observation guessing NUM_BEDROOMS = 3.for 1800SQ_FT
property_data[8:]['NUM_BEDROOMS'] = 3.0
```

# Chi-Square test

```python
table = pd.crosstab(property_data['ST_NAME'],property_data['OWN_OCCUPIED'])
table
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(table)
```

p =  0.061099480960332735 which is greater than 0.05 hence, there is no asssociation between two variable
but most the cell data is < 5 hence Chi-Square Test is not reliable

# Using researchpy library

# Chi-Square Test

```python
import researchpy as rp
```

```python
table, results = rp.crosstab(property_data['ST_NAME'],property_data['OWN_OCCUPIED'],test='chi-square')
    
print(table)
results
```

# Fisher's Exact Test

```python
table, results = rp.crosstab(property_data['ST_NAME'],property_data['OWN_OCCUPIED'],test='fisher')
    
print(table)
results
```

in jupyter fisher exact test can not be performed for sizes greater than (2,2)

# g-test

```python
table, results = rp.crosstab(property_data['ST_NAME'],property_data['OWN_OCCUPIED'],test='g-test')
    
print(table)
results
```

p-value = 0.0490
hence according to g-test the two variables are related to each other
