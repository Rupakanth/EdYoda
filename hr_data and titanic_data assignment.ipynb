convert avg montly hours into days and hours in hr_data

```python
def f(r):
    return (str(r.average_montly_hours//24) + "days" + str(r.average_montly_hours%24) + "hours")

hr_data['avg_working_hours']=hr_data.apply(f,axis = 1)
hr_data.head()
```

Give hike to people still with company & did in top 25%ile of performance

```python
def f(r):
    if (r.last_evaluation > 0.87) & (r.left ==0) & (r.salary == 'low'):
        return "medium"
    else:
        return r.salary
        


hr_data["top_performers_hike"]=hr_data.apply(f,axis = 1)
```

```python
hr_data[(hr_data.last_evaluation > 0.87) & (hr_data.salary != hr_data.top_performers_hike)].head()
```

Create a new column with values KID,ADULT & OLD based on age

```python
filling missing age values with the mean of age
```

```python
titanic_data.Age.mean()
```

```python
titanic_data.fillna({'Age':29.7},inplace = True)
```

```python
def f(r):
    if r.Age <= 20:
        return "KID"
    elif r.Age <= 40:
        return "ADULT"
    else:
        return "OLD"

titanic_data['Age_category'] = titanic_data.apply(f,axis = 1)
```
